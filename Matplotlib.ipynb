```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
#importing the dataset

df = pd.read_csv('Customer_data_updated.csv')
df
```

```python
df.dtypes
```

```python
df["Total Claim Amount"]=df["Total Claim Amount"].str.replace('[\$\,]|\.\d*','').astype(float)
```

```python
df.dtypes
```

```python
#Appling Cumulative summation
df["Total Number of Policies"]=df["Number of Policies"].cumsum().astype(int)
df
```

```python
# Adding date column
df["Policy Date"]=pd.date_range("25/3/2018",periods=len(df))
df
```

```python
# different styles
plt.style.available
```

# Line Plot

```python
df['Income'].plot()

plt.title('Income Info')
plt.xlabel('x-axis')
plt.ylabel('Income')

plt.savefig('Figures/Income Info.png')
```

# Histogram

```python

x = df['Number of Policies']
y = df['Total Claim Amount']

fig,ax = plt.subplots(figsize=(6,7))
ax.hist(x, color = '#a9bca9')


plt.title('Total claim amout by number of policies')
plt.xlabel('Number of Policies')
plt.ylabel('Total Claim Amount')
plt.show()

plt.savefig('Figures/Total claim amout by number of policies.png')
```

# Scatter Plot

```python
df.plot(x = 'Total Claim Amount', 
        y = 'Income', 
        kind = 'scatter', 
        color = '#f88f01',
        title = 'Performance based Makret place value for players')
plt.savefig('Figures/Performance based Makret place value for players.png')
```

# Bar Plot

```python
df.plot(x = 'Gender', 
        y = 'Total Claim Amount', 
        kind = 'bar', 
        color = '#f25287',
        title = 'Total Claim Amount by Gender')
plt.savefig('Figures/Total Claim Amount by Gender.png')
```

# Bar Plot Horizontal

```python
df.plot(x = 'State', 
        y = 'Income', 
        kind = 'barh', 
        color ='LightSeaGreen' ,
        title = 'Income By State')
plt.savefig('Figures/Income By State.png')
```

# Histogram

```python
x = df['Total Claim Amount']
y = df['Income']

fig,ax = plt.subplots(figsize=(6,5))
ax.hist(x)


plt.title('Total Claim Amount by Income')
plt.xlabel('Total Claim Amount')
plt.ylabel('Income')
plt.style.use('seaborn-muted')
plt.show()

plt.savefig('Figures/Total Claim Amount by Income.png')
```

```python
#Manipulating Data
df["State"]=df["State"].str.lower()
df
```

```python
##filling the missing data
df["Months Since Last Claim"].fillna(df["Months Since Last Claim"].mean(),inplace=True)
df
```

```python
##filling the missing data
df["Income"].fillna(df["Income"].mean(),inplace=True)
df
```

```python
#Creating a new column using the existing data
Number_of_Open_Complaints=pd.Series([2,0,1,2,4,1,0,3,5,2,2,0,0,1,2,0,3,3,5,2,0,3,1])
df["Complaints"]=Number_of_Open_Complaints
df
```

```python
df["Complaints"].fillna(2,inplace=True)
df
```

```python
#creating a column using python list
Customer_Gender=['F','F','M','F','M','M','M','M','F','F','M','M','M','M','F','F','F','M','M','M','M','M','M','F','F','F','F','M','M']
df["Gender"]=Customer_Gender
df
```

```python
#Removing a column
df.drop("Gender",axis=1)
```
